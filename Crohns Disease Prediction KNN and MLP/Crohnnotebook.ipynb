```python
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

df = pd.read_csv("CrohnD.csv")
```

```python
df = df.drop(['Unnamed: 0', 'ID'], axis = 1)
```

```python
df
```

```python
df['country'] = df['country'].replace(['c1'], '0')
df['country'] = df['country'].replace(['c2'], '1')
df['sex'] = df['sex'].replace(['F'], '0')
df['sex'] = df['sex'].replace(['M'], '1')
```

```python
X = df.drop(['treat'], axis = 1)
y = df['treat']
```

```python
model = MLPClassifier(hidden_layer_sizes=(15,),max_iter=10000)
scores = cross_val_score(model,X,y,cv = 5)
print("Fold Accuracies: {}".format(scores))
print("Accuracy: {}".format(scores.mean()))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
modelKNN = KNeighborsClassifier(n_neighbors=5)
train_X, test_X,train_y,test_y = train_test_split(X,y,train_size=0.8)
modelKNN.fit(train_X, train_y)
```

```python
predict_y = modelKNN.predict(test_X)
print("Train-Test Accurcay: {}".format(accuracy_score(test_y,predict_y)))
```
