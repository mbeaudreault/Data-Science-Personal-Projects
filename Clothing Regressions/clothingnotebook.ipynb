```python
The questions that I am trying to answer in my analysis is to find general
optimization of clothing stores. For example, how many workers should you have, 
how many hours should they work, how big should your store be, and how much
should you invest into your own store. In order to answer these question I plan 
to use regressions to determine what attributes produce the greatest
values for margins and sales. I can also use the clustering method to idenfity
different types of successful stores and how those work in relation to others.
I would like to build models that model which company attributes result in a 
positive margin and which produce negative. I expect there to be strong 
correlations between things like number of owners, start time, and number of
number of workers and sales or margins. Because the reason a store is that old
or has that many owners is becaues they were already successful. In order to 
counter act this bias I can create ratio to compare to, for example I could
create a dimension like number of full time workers/size. I would expect every 
variables relation to magirn and sales to look similar to a distribution curve.
In that there is a build up where for example a greater number of workers results
in greater sales up until a point where the company gets deminishing returns.
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
```

```python
Z = pd.DataFrame(df["start"])
L = pd.DataFrame(df["margin"])
z = df["start"]
l = df["margin"]
model = KNeighborsRegressor(n_neighbors = 3)
model.fit(Z,L)
print("R^2 score: {}".format(model.score(Z,L)))
xfit = pd.DataFrame([i for i in range(-1, 100)])
yfit = model.predict(xfit)
plt.scatter(z,l)
plt.plot(xfit,yfit, color = 'red')
plt.show()
```

```python
This reccurance had one of the strongest r^2 score of all of the reccuraces I 
have tested. There is a very strong relation between the year the company was 
created and the gross profit margin. This relationship has a very strong positive
correlation followed what appears to be deminishin returns of a weaker 
correlation. This information makes sense and is a pretty good example of the
difference between causation and correlation. If a fashion store is struggling 
it would not make sense to tell the owner to hand in there because his or her
margins will go up. Instead this data probably represents that fact that the
company has lasted this long because they have good margins. What is interesting 
is how strong the correlation of the beginning part of the graph, it is almost
a perfectly straight line. Eventually companies appear to have deminishing returns
on their margins as they grow older. I think this could be the result of older
companies possible selling more and caring less about their margins because their
number of sales will make up for that. Or the bigger companies have a harder time
streamlining itself resulting in issues.
```

```python
X = pd.DataFrame(df["nfull"])
Y = pd.DataFrame(df["tsales"])
x = df["nfull"]
y = df["tsales"]
model = KNeighborsRegressor(n_neighbors = 3)
model.fit(X,Y)
print("R^2 score: {}".format(model.score(X,Y)))
xfit = pd.DataFrame([i for i in range(-1, 10)])
yfit = model.predict(xfit)
plt.scatter(x,y)
plt.plot(xfit,yfit, color = 'red')
plt.show()
```

```python
This analyic compares the number of full time workers to total sales
and is in a similar situation as the one above. That being a question of
causation or correlation. However; this one is definatily less cut and dry. 
While a company who sells more will be able to hire more worker. It also makes 
sense that a company who highers more employees would sell more clothes. This 
relationship appears to be accelerating as more workers are added. It is 
important to note there is not the greatest variety in the number of full time
workers and that might have an adverse affect on the realationship.
```

```python
X = pd.DataFrame(df["hourspw"])
Y = pd.DataFrame(df["margin"])
x = df["hourspw"]
y = df["margin"]
model = KNeighborsRegressor(n_neighbors = 3)
model.fit(X,Y)
print("R^2 score: {}".format(model.score(X,Y)))
xfit = pd.DataFrame([i for i in range(-1, 40)])
yfit = model.predict(xfit)
plt.scatter(x,y)
plt.plot(xfit,yfit, color = 'red')
plt.show()
```

```python
This final reccurance denotes the relationship between hours worked per worker 
and margin. This analytic was particularly interesting to me because it can offer
information as to whether or not is it better to have fewer wokers with more 
experience or more workers with less expiernce. In general it seems that as 
workers gain more experience the company has greater margins, although the r^2 
value is not very strong.
```
