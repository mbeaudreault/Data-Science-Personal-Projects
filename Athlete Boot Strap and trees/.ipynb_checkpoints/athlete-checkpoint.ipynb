```python
import pandas as pd
from sklearn import tree
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_csv("athletedataset.csv")
X = df.drop(['sex','sport', 'Unnamed: 0'], axis = 1)
y = df['sport']
```

```python
X
```

```python
model = tree.DecisionTreeClassifier(criterion='entropy', max_depth=2)
```

```python
scores = cross_val_score(model,X,y,cv=3)
print("Fold Accuracies: {}".format(scores))
print("Accuracy: {}".format(scores.mean()))
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from treeviz import tree_print
```

```python
model = model = tree.DecisionTreeClassifier()
param_grid = {'max_depth': list(range(1,11)),
              'criterion': ['entropy', 'gini']
              }
grid = GridSearchCV(model, param_grid, cv=3)

 
grid.fit(X,y)

print("Best parameters: {}".format(grid.best_params_))


print("Best tree:")
tree_print(grid.best_estimator_,X)

        
predict_y = grid.best_estimator_.predict(X)
actual_y = y
        
print("Accuracy: {}".format(accuracy_score(actual_y, predict_y)))
```

```python
from sklearn.model_selection import train_test_split
from numpy import percentile
```

```python
def bootstrap(model,D,target_name):
    rows,__ = D.shape
    acc_list = []
    for i in range(200):
        B = D.sample(n=rows,replace=True)
        X = B.drop(target_name,1)
        y = B[target_name]
        train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=0.8)
        model.fit(train_X, train_y)
        predict_y = model.predict(test_X)
        acc_list.append(accuracy_score(test_y, predict_y))
    acc_list.sort()
    ub = percentile(acc_list,97.5)
    lb = percentile(acc_list,2.5)
    return (lb, ub)
```

```python
Z = df.drop(['sex', 'Unnamed: 0'], axis = 1)
```

```python
t1 = tree.DecisionTreeClassifier(criterion='entropy', max_depth=3)
t2 = tree.DecisionTreeClassifier(criterion='entropy', max_depth=None)

print("Confidence interval max_depth=3: {}".format(bootstrap(t1,Z,'sport')))
print("Confidence interval max_depth=None: {}".format(bootstrap(t2,Z,'sport')))
```
